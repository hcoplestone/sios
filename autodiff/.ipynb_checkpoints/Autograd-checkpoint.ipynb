# Autograd

First, let's import autograd and numpy. 

```python
from math import *
import autograd.numpy as np
from autograd import grad
from autograd import elementwise_grad as egrad  # for functions that vectorize over inputs
import matplotlib.pyplot as plt
```

Define $$f(x) = sin(x)$$

```python
def f(x):
    return np.sin(x)
```

```python
f(pi/2)
```

We expect
$$f'(x) = cos(x)$$

Use autograd to numerically calculate the derivative and work out the residuals for $x \in [0, 2\pi]$ to test accuracy.

```python
grad_f = egrad(f)
```

```python
def actual_grad_f(x):
    return np.cos(x)
```

Let's plot everything to see if we're getting the right vibe.

```python
x = np.linspace(0,2*pi)
```

```python
fig = plt.figure()
result = f(x)
plt.plot(x, result)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('y = f(x)');
```

```python
fig = plt.figure()

plt.subplot(2,2,1)
plt.title('Numerical gradient')
plt.plot(x, grad_f(x))
plt.xlabel('$x$')

plt.subplot(2,2,2)
plt.title('Actual gradient')
plt.plot(x, actual_grad_f(x))
plt.ylabel('$f\'(x)$')
plt.xlabel('$x$')
```

```python
residual = grad_f(x) - actual_grad_f(x)
```

```python
fig = plt.figure()
plt.title('Residual')
plt.plot(x, residual)
```
